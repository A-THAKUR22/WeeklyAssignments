A simple command-line quiz game that reads questions from a file, allows users to answer them, and provides a score at the end. The questions and answers should be stored in a text file.

```python
import random
import sys
class Question:
    def __init__(self, question_text, options, correct_answer):
        self.question_text = question_text
        self.correct_answer = correct_answer
        self.options = options
#Check whether the answer is correct or not
    def check_answer(self, user_answer):
        return user_answer.lower() == self.correct_answer.lower()
#Display the structures of storing the questions,options and the answers
    def printquestion(self):
        print(self.question_text)
        print(self.options)
        print(self.correct_answer)

#Read the lines in file and store the questions and their options with their respective answers.
def read_questions(filename):
    questions = []

    try:#Check whether the file exists or not.
      with open(filename, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        question_text = ''
        options = {}
        correct_answer = ''
        for line in lines:
          line = line.strip()
          if line.startswith("Q"):#Check for adding questions and storing them in a list
                if question_text:
                    question = Question(question_text, options, correct_answer)
                    questions.append(question)

                    options = {}
                question_text = line.split('.', 1)[1].strip()
                #Add options of MCQ type questions in the dictionary
          elif len(line)>2 and line[1]=="." and line[0].isalpha():
                option_key = line[0]
                option_value = line[3:].strip()
                options[option_key] = option_value
          elif line == "":#Skip if any empty line is encountered
                continue
          elif line.startswith('Answer:'):#Store the correct answers
                correct_answer = line.split(': ')[1]
        if question_text:#Add the last question
            question = Question(question_text, options, correct_answer)
            questions.append(question)

    except FileNotFoundError:
      print("The file doesn't exist!!")
      sys.exit(1)


    return questions


#Handle the quiz game and calculate the scores while dealing with negative marking
def quiz_game(questions):
    score = 0
    mcq_ques=[]
    correct_mcq=0
    incorrect_mcq=0

    total_ques=len(questions)
    random.shuffle(questions)
    for idx, question in enumerate(questions):
        print(f"Question {idx + 1}: {question.question_text}")


        if question.options:
          for key, value in question.options.items():
            print(f"{key}: {value}")
            mcq_ques.append(key)
          while True:
            user_answer = input("Your answer: ").strip().lower()
            if user_answer in mcq_ques:

                break
            else:
              print("Invalid option try again")
        else:
          try:
                user_answer = input("Your answer: ").strip().lower()
          except KeyboardInterrupt:
                print("The quiz has been interrupted by the user.")
                sys.exit(1)
          except ValueError:
                print("An error has occurred.")





        if question.check_answer(user_answer):
            print("Correct!\n")

            if question.options:#If the question is mcq then and is correct add one mark
              correct_mcq+=1
              score += 1
            else:#If the question is not a mcq type and is correct add two marks
              score+=2

        else:
          print(f"Incorrect! Correct Answer: {question.correct_answer}.\n")
          if question.options:#If the incorrect question is a mcq type increment the count of incorrect questions
            incorrect_mcq+=1
            if incorrect_mcq==3:#If 3 mcq questions are incorrect we deduct one mark from the total score and reset the count of incorrect questions
              score-=1
              incorrect_mcq=0

        print(f"Your score is :{score}")


questions = read_questions('NewMCQ Quiz.txt')
quiz_game(questions)
```
